**Answer:**


I frame my lane as a **ranking and scoring** machine learning task. The goal is to prioritize web pages based on their likelihood of needing a content review. ## 2. Target or proxy

==================================================================

**Answer**

The target of my project is the review priority of each web page. Since the dataset does not contain a direct priority label, I will use a proxy label based on observed performance signals, such as declining trends and other available metrics. ## 3. Success metric

==================================================================

**Answer**


The success metric for my project is **Precision@50**. This metric measures how many of the top 50 recommended pages actually deserve review. A higher Precision@50 means the model is better at identifying the most important pages, helping the SEO team use their limited review time more effectively.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

==================================================================

**Answer**

A simple rule cannot capture the complex relationships between multiple performance signals. Page performance depends on many factors, such as clicks, impressions, CTR, engagement, and trend, which interact in different ways. Machine learning can learn these patterns from historical data and provide more accurate prioritization than a fixed rule or a single if-statement.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**=====================================================================**

**Answer:**

**Decision of my work improves:** Which pages should be reviewed or refreshed first when the reviewer has limited time and can only inspect a small number of pages (for example, the top 20–50 pages).

**Who acts on it:** A content editor or SEO reviewer who manually evaluates the recommended pages and decides whether to refresh, expand, optimize, or leave the content unchanged.

**Cost of a wrong recommendation:**
- **False Positive:** A page is incorrectly identified as high priority. This wastes the reviewer's limited time and may delay attention to pages that actually need improvement.
- **False Negative:** A page that genuinely needs attention is not identified. As a result, it may continue losing visibility and traffic before the issue is noticed.

```python
import pandas as pd

# Load the downloaded CSV file into a DataFrame
df = pd.read_csv("content_refresh_anonymized.csv")

# Display the shape of the DataFrame
print(f"DataFrame shape: {df.shape}")

# Display the first few rows to confirm loading
display(df.head())
```

```python
df.info()
```

```python
print("Total unique clients:", df['client_id'].nunique())

# Apply the guide's minimum filters
filtered = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)]
print("\n filtered shape:",filtered.shape)


# Number 1: how much of the data shows current decline
down_trend_count = df['trend_direction'].value_counts().get('down', 0)
total_rows = len(df)
down_pct = (down_trend_count / total_rows) * 100

print(f"\n Total 'down' trends: {down_trend_count}")
print(f"Percentage declining: {down_pct:.1f}%")


 # Number 2: how many pages have real visibility worth reviewing
visible_pages = (filtered['impressions_90d'] >= 500).sum()
print("\n Pages with impressions_90d >= 500:", visible_pages)
```
